```python
import requests

reqUrl = "https://internshala.com/internships/page-1/"

headersList = {
 "Accept": "*/*",
 "User-Agent": "Thunder Client (https://www.thunderclient.com)" 
}

payload = ""

response = requests.request("GET", reqUrl, data=payload,  headers=headersList)

soup = BeautifulSoup(response.content, 'html.parser')
```

```python

nested_divs = soup.find_all('div', class_='container-fluid individual_internship visibilityTrackerItem')

for target_div in nested_divs:

        try:
               internship_title =target_div.select_one('.individual_internship_header .heading_4_5 a').text.strip()
        except:
              internship_title = "None"
        
        try:
            company_name =target_div.select_one('.company_name a').text.strip()
        except:
              company_name = "None"
        
        try:
                location =target_div.select_one('#location_names span a').text.strip()
        except:
              location = "None"
        
        try:
            duration = soup.select_one('.other_detail_item_row .other_detail_item.large_stipend_text div.item_body').text.strip()
        except:
              duration = "None"
        
        try:
            stipend = target_div.select_one('.stipend_container .item_body').text.strip()
        except:
              stipend = "None"

        start_date_element = target_div.select_one('#start-date-first .start_immediately_desktop')
        start_date = start_date_element.text.strip() if start_date_element else 'Not specified'

        stipend_element = target_div.select_one('.stipend_container .item_body')
        stipend = stipend_element.text.strip() if stipend_element else 'Not specified'

        # Print the extracted data
        print("Internship Title:", internship_title)
        print("Company Name:", company_name)
        print("Location:", location)
        print("Start Date:", start_date)
        print("Duration:", duration)
        print("Stipend:", stipend)

```

```python
import psycopg2
import requests
from psycopg2 import sql

def push_data(data, conn):
            insert_query = sql.SQL("""
                  INSERT INTO {} 
                  (source_name, author, title, description, url, url_to_image, published_at, content, category)
                  VALUES 
                  (%(source_id)s, %(source_name)s, %(author)s, %(title)s, %(description)s, %(url)s, %(urlToImage)s, %(publishedAt)s, %(content)s, %(category)s)
            """).format(sql.Identifier(table_name))

            with conn.cursor() as cur:
                  cur.execute(insert_query, data)

            conn.commit()

def connect_to_database(dbname, user, password, host, port):
        # Establish a connection
        conn = psycopg2.connect(dbname=dbname, user=user, password=password, host=host, port=port)

        # Create a cursor object to execute SQL queries
        return conn


def push_data(data, conn, table_name):
    insert_query = sql.SQL("""
        INSERT INTO {} 
        (Internship_Title, Company_Name, Location, Start_Date, Duration, Stipend)
        VALUES 
        (%(Internship_Title)s, %(Company_Name)s, %(Location)s, %(Start_Date)s, %(Duration)s, %(Stipend)s)
    """).format(sql.Identifier(table_name))

    # Create a cursor and execute the insert query
    with conn.cursor() as cursor:
        cursor.execute(insert_query, data)

    # Commit the changes to the database
    conn.commit()

conn = connect_to_database(dbname, user, password, host, port)
def process(num):

    reqUrl = "https://internshala.com/internships/page-{}/".format(num)

    headersList = {
    "Accept": "*/*",
    "User-Agent": "Thunder Client (https://www.thunderclient.com)" 
    }

    payload = ""

    response = requests.request("GET", reqUrl, data=payload,  headers=headersList)

    soup = BeautifulSoup(response.content, 'html.parser')

    nested_divs = soup.find_all('div', class_='container-fluid individual_internship visibilityTrackerItem')

    for target_div in nested_divs:

        try:
               internship_title =target_div.select_one('.individual_internship_header .heading_4_5 a').text.strip()
        except:
              internship_title = "None"
        
        try:
            company_name =target_div.select_one('.company_name a').text.strip()
        except:
              company_name = "None"
        
        try:
                location =target_div.select_one('#location_names span a').text.strip()
        except:
              location = "None"
        
        try:
            duration = soup.select_one('.other_detail_item_row .other_detail_item:nth-child(2) .item_body').text.strip()
        except:
              duration = "None"
        
        try:
            stipend = target_div.select_one('.stipend_container .item_body').text.strip()
        except:
              stipend = "None"

        start_date_element = target_div.select_one('#start-date-first .start_immediately_desktop')
        start_date = start_date_element.text.strip() if start_date_element else 'Not specified'

        stipend_element = target_div.select_one('.stipend_container .item_body')
        stipend = stipend_element.text.strip() if stipend_element else 'Not specified'

        data_dict = {
            'Internship_Title': internship_title,
            'Company_Name': company_name,
            'Location': location,
            'Start_Date': start_date,
            'Duration': duration,
            'Stipend': stipend,
            }
        # Assuming conn is your database connection
        push_data(data_dict, conn, 'internship')
```

```python
import pandas as pd

data = pd.read_csv("data/internship.csv")
data.head()
```

```python
data["stipend"] = data['stipend'].apply(lambda x: '0' if 'Unpaid' in str(x) else x)
data["stipend"] = data["stipend"].str.replace("/month", "")
data["stipend"] = data["stipend"].str.replace("lump sum", "")
data["min stipend"] = data["stipend"].str.split("-").str[0].str.strip().str.replace("₹", "")
data["max stipend"] = data["stipend"].str.split("-").str[-1].str.strip().str.replace("₹", "")
data["Other Benifits"] = data["stipend"].str.split('+').apply(lambda x: x[-1].strip() if len(x) > 1 else None)
data["stipend"] = data["stipend"].str.split('+').str[0].str.strip()
```
