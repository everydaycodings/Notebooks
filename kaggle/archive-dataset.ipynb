```python
import psycopg2,re
import requests
from psycopg2 import sql
from bs4 import BeautifulSoup

dbname = "postgres"
user = ""
password = ""
host = "n"
port = 5432
table_name = ""


def connect_to_database(dbname, user, password, host, port):
        # Establish a connection
        conn = psycopg2.connect(dbname=dbname, user=user, password=password, host=host, port=port)

        # Create a cursor object to execute SQL queries
        return conn


def push_data(data, conn, table_name):
    insert_query = sql.SQL("""INSERT INTO {} (Identifier, Link, Title, Description, Subject, Collection, MediaType, ItemSize, Downloads,PublicDate, Creator, Contributor, Format, License) VALUES 
                           (%(Identifier)s, %(Link)s, %(Title)s, %(Description)s, %(Subject)s, %(Collection)s, %(MediaType)s, %(ItemSize)s, %(Downloads)s, %(PublicDate)s, %(Creator)s, %(Contributor)s, %(Format)s, %(License)s)
                           """).format(sql.Identifier(table_name))

    with conn.cursor() as cursor:
        cursor.execute(insert_query, data)

    # Commit the changes to the database
    conn.commit()

conn = connect_to_database(dbname, user, password, host, port)
def process(num):

    reqUrl = "https://archive.org/advancedsearch.php?q=mediatype%3A%28movies%29&fl%5B%5D=avg_rating&fl%5B%5D=collection&fl%5B%5D=contributor&fl%5B%5D=creator&fl%5B%5D=description&fl%5B%5D=downloads&fl%5B%5D=format&fl%5B%5D=genre&fl%5B%5D=identifier&fl%5B%5D=item_size&fl%5B%5D=language&fl%5B%5D=licenseurl&fl%5B%5D=mediatype&fl%5B%5D=num_reviews&fl%5B%5D=publicdate&fl%5B%5D=publisher&fl%5B%5D=subject&fl%5B%5D=title&sort%5B%5D=&sort%5B%5D=&sort%5B%5D=&rows=10000&page={}&output=json&callback=callback&save=yes".format(num)

    headersList = {
    "Accept": "*/*",
    "User-Agent": "Thunder Client (https://www.thunderclient.com)" 
    }

    payload = ""

    response =requests.request("GET", reqUrl, data=payload,  headers=headersList).text

    pattern = re.compile(r'\{.*\}')

    # Find the JSON part using the regular expression
    match = pattern.search(response)

    # Extract the matched JSON string
    if match:
        json_part = match.group(0)
        
        # Parsing the JSON
        results = json.loads(json_part)["response"]["docs"]

        for data in results:

            try:
                identifier = data["identifier"]
            except:
                identifier = "None"
                
            try:
                link = f"https://archive.org/details/{identifier}"
            except:
                link = "None"
            try:
                title = data["title"]
            except:
                title = "None"
            try:
                description = data["description"]
            except:
                description = "None"
            try:
                subject = data["subject"]
            except:
                subject = "None"
            try:
                collection = data["collection"]
            except:
                collection = "None"
            try:
                mediatype = data["mediatype"]
            except:
                mediatype = "None"
            try:
                item_size = data["item_size"]
            except:
                item_size = "None"
            try:
                downloads = data["downloads"]
            except:
                downloads = "None"
            try:
                publicdate = data["publicdate"]
            except:
                publicdate = "None"
            try:
                creator = data["creator"]
            except:
                creator = "None"
            try:
                contributor = data["contributor"]
            except:
                contributor = "None"
            try:
                mformat = data["format"]
            except:
                mformat = "None"
            try:
                licence = data["licenseurl"]
            except:
                licence = "None" 



            data_dict = {
                'Identifier': identifier,
                'Link': link,
                'Title': title,
                'Description': description,
                'Subject': subject,
                'Collection': collection,
                'MediaType': mediatype,
                'ItemSize': item_size,
                'Downloads': downloads,
                'PublicDate': publicdate,
                'Creator': creator,
                'Contributor': contributor,
                'Format': mformat,
                'License': licence
                }
            # Assuming conn is your database connection
            push_data(data_dict, conn, table_name)
        
        return len(results)
```

```python
def append_word_to_file(word, filename="output.log"):
    with open(filename, 'a') as file:
        file.write(str(word) + '\n')

for i in range(1, 100):
    length = process(i)
    append_word_to_file("{} - {}".format(i, length))
    print(i)
```
