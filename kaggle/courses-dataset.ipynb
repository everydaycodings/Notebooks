```python
import psycopg2
import requests
from psycopg2 import sql


def connect_to_database(dbname, user, password, host, port):
        # Establish a connection
        conn = psycopg2.connect(dbname=dbname, user=user, password=password, host=host, port=port)

        # Create a cursor object to execute SQL queries
        return conn


def push_data(data, conn, table_name):
    insert_query = sql.SQL("""INSERT INTO {} (Partner, Course, Skills, Rating, ReviewCount, Level, CertificateType, Duration, CreditEligibility) VALUES 
                           (%(Partner)s, %(Course)s, %(Skills)s, %(Rating)s, %(ReviewCount)s, %(Level)s, %(CertificateType)s, %(Duration)s, %(CreditEligibility)s)
                           """).format(sql.Identifier(table_name))

    with conn.cursor() as cursor:
        cursor.execute(insert_query, data)

    # Commit the changes to the database
    conn.commit()

conn = connect_to_database(dbname, user, password, host, port)
def process(num):

    reqUrl = "https://www.coursera.org/search?page={}".format(num)

    headersList = {
    "Accept": "*/*",
    "User-Agent": "Thunder Client (https://www.thunderclient.com)" 
    }

    payload = ""

    response = requests.request("GET", reqUrl, data=payload,  headers=headersList)

    soup = BeautifulSoup(response.content, 'html.parser')

    list_items = soup.find('ul', class_='cds-9').find_all('li')

    for soup in list_items:
        
        try:
            partner_name = soup.find('p', class_='cds-119 cds-ProductCard-partnerNames css-dmxkm1 cds-121').text
        except:
             partner_name = "None"
        
        try:
            title = soup.find('h3', class_='cds-119 cds-CommonCard-title css-e7lgfl cds-121').text
        except:
             title = "None"
        
        try:
            skills = soup.find('p', class_='cds-119 cds-Typography-base css-dmxkm1 cds-121').text.replace("Skills you'll gain:", "").split(",")
        except:
             skills = "None"
        
        try:
            rating = soup.find('p', class_='cds-119 css-11uuo4b cds-121').text
        except:
             rating = "None"
        
        try:
            review_count = soup.find('div', class_='product-reviews').text.strip().split("(")[-1].replace(" reviews)", "") # Extracting the review count
        except:
             review_count = "None"
        
        try:
            metadata = soup.find('div', class_='cds-CommonCard-metadata').text.strip().split("·")
            try:
                level = metadata[0]
            except:
                 level = "None"
            try:
                certificat_type = metadata[1]
            except:
                 certificat_type = "None"
            try:
                duration = metadata[2]
            except:
                 duration = "None"

        except:
             level = "None"
             certificat_type = "None"
             duration = "None"
        
        try:
            badge_div = soup.find('div', {'data-e2e': 'rc-CreditEligibleBadge'})
            credit_eligible_text = badge_div.find('p', {'data-e2e': 'creditEligibleText'}).text.strip() if badge_div else None
            credit_eligible = bool(credit_eligible_text)
        except:
             credit_eligible = "None"


        data_dict = {
            'Partner': partner_name,
            'Course': title,
            'Skills': skills,
            'Rating': rating,
            'ReviewCount': review_count,
            'Level': level,
            'CertificateType': certificat_type,
            'Duration': duration,
            'CreditEligibility': credit_eligible
            }
        # Assuming conn is your database connection
        push_data(data_dict, conn, table_name)
```

```python
import psycopg2
import requests
from psycopg2 import sql
from selenium import webdriver
import time


dbname = "postgres"
user = ""
password = ""
host = ""
port = 5432
table_name = "edx"


def deep_dive(url):

    # Set up the WebDriver (make sure to have the appropriate webdriver installed)
    driver = webdriver.Chrome()

    # Open the URL
    driver.get(url)
    # Get the page source after JavaScript execution
    html_code = driver.page_source
    soup1 = BeautifulSoup(html_code, 'html.parser')
    driver.quit()

    soup = soup1.find("div", class_="at-a-glance")

    li_list = soup.find_all('li')

    info_dict = {}

    # Extract and store key-value pairs in the dictionary
    for li in li_list:
        li_text = li.get_text(strip=True)
        # Split the text at the ':' character
        if ':' in li_text:
            key, value = li_text.split(':', 1)
            # Remove any leading or trailing whitespace from the key and value
            key = key.strip()
            value = value.strip()
            # Store the key-value pair in the dictionary
            info_dict[key] = value
    
    return info_dict


def connect_to_database(dbname, user, password, host, port):
        # Establish a connection
        conn = psycopg2.connect(dbname=dbname, user=user, password=password, host=host, port=port)

        # Create a cursor object to execute SQL queries
        return conn


def push_data(data, conn, table_name):
    insert_query = sql.SQL("""INSERT INTO {} (Title, Link, Institution, Subject, Level, Prerequisites, Language, VideoTranscript, AssociatedPrograms, AssociatedSkills) VALUES 
                           (%(Title)s, %(Link)s, %(Institution)s, %(Subject)s, %(Level)s, %(Prerequisites)s, %(Language)s, %(VideoTranscript)s, %(AssociatedPrograms)s, %(AssociatedSkills)s)
                           """).format(sql.Identifier(table_name))

    with conn.cursor() as cursor:
        cursor.execute(insert_query, data)

    # Commit the changes to the database
    conn.commit()

conn = connect_to_database(dbname, user, password, host, port)
def process4(num):

    url = "https://www.edx.org/search?tab=course&page={}".format(num)

    # Set up the WebDriver (make sure to have the appropriate webdriver installed)
    driver = webdriver.Chrome()

    # Open the URL
    driver.get(url)

    time.sleep(4)
    # Get the page source after JavaScript execution
    html_code = driver.page_source

    driver.quit()

    soup = BeautifulSoup(html_code, 'html.parser')
    
    container_div = soup.find(attrs={'data-testid': 'card-container'})
    course_divs = container_div.find_all('div', class_='base-card-wrapper')

    for soup in course_divs:
        
        try:
            course_title = soup.select_one('.pgn__card-header-title-md span span').text
        except:
             course_title = "None"
        
        try:
            organization_name = soup.select_one('.pgn__card-header-subtitle-md span span').text
        except:
             organization_name = "None"
             
        try:
            course_url = "https://www.edx.org{}".format(soup.select_one('.base-card-link')['href'].split("?")[0])
        except:
             course_url = "None" 

        try:
            deep = deep_dive(course_url)
        except:
            deep = "None"

        try:
            Subject = deep["Subject"]
        except:
            Subject = "None"
        try:
            Level = deep["Level"]
        except:
            Level = "None"
        try:
            Prerequisites = deep["Prerequisites"]
        except:
            Prerequisites = "None"
        try:
            Language = deep["Language"]
        except:
            Language = "None"
        try:
            VideoTranscript = deep["Video Transcript"]
        except:
            VideoTranscript = "None"
        try:
            AssociatedPrograms = deep["Associated programs"]
        except:
            AssociatedPrograms = "None"
        try:
            AssociatedSkills = deep["Associated skills"]
        except:
            AssociatedSkills = "None"

        data_dict = {
            'Title': course_title,
            'Link': course_url,
            'Institution': organization_name,
            'Subject': Subject,
            'Level': Level,
            'Prerequisites': Prerequisites,
            'Language': Language,
            'VideoTranscript': VideoTranscript,
            'AssociatedPrograms': AssociatedPrograms,
            'AssociatedSkills': AssociatedSkills,
            }
        
        # Assuming conn is your database connection
        push_data(data_dict, conn, table_name)
```
