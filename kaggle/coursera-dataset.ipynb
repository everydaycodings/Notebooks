```python
import psycopg2
import requests
from psycopg2 import sql


def connect_to_database(dbname, user, password, host, port):
        # Establish a connection
        conn = psycopg2.connect(dbname=dbname, user=user, password=password, host=host, port=port)

        # Create a cursor object to execute SQL queries
        return conn


def push_data(data, conn, table_name):
    insert_query = sql.SQL("""INSERT INTO {} (Partner, Course, Skills, Rating, ReviewCount, Level, CertificateType, Duration, CreditEligibility) VALUES 
                           (%(Partner)s, %(Course)s, %(Skills)s, %(Rating)s, %(ReviewCount)s, %(Level)s, %(CertificateType)s, %(Duration)s, %(CreditEligibility)s)
                           """).format(sql.Identifier(table_name))

    with conn.cursor() as cursor:
        cursor.execute(insert_query, data)

    # Commit the changes to the database
    conn.commit()

conn = connect_to_database(dbname, user, password, host, port)
def process(num):

    reqUrl = "https://www.coursera.org/search?page={}".format(num)

    headersList = {
    "Accept": "*/*",
    "User-Agent": "Thunder Client (https://www.thunderclient.com)" 
    }

    payload = ""

    response = requests.request("GET", reqUrl, data=payload,  headers=headersList)

    soup = BeautifulSoup(response.content, 'html.parser')

    list_items = soup.find('ul', class_='cds-9').find_all('li')

    for soup in list_items:
        
        try:
            partner_name = soup.find('p', class_='cds-119 cds-ProductCard-partnerNames css-dmxkm1 cds-121').text
        except:
             partner_name = "None"
        
        try:
            title = soup.find('h3', class_='cds-119 cds-CommonCard-title css-e7lgfl cds-121').text
        except:
             title = "None"
        
        try:
            skills = soup.find('p', class_='cds-119 cds-Typography-base css-dmxkm1 cds-121').text.replace("Skills you'll gain:", "").split(",")
        except:
             skills = "None"
        
        try:
            rating = soup.find('p', class_='cds-119 css-11uuo4b cds-121').text
        except:
             rating = "None"
        
        try:
            review_count = soup.find('div', class_='product-reviews').text.strip().split("(")[-1].replace(" reviews)", "") # Extracting the review count
        except:
             review_count = "None"
        
        try:
            metadata = soup.find('div', class_='cds-CommonCard-metadata').text.strip().split("·")
            try:
                level = metadata[0]
            except:
                 level = "None"
            try:
                certificat_type = metadata[1]
            except:
                 certificat_type = "None"
            try:
                duration = metadata[2]
            except:
                 duration = "None"

        except:
             level = "None"
             certificat_type = "None"
             duration = "None"
        
        try:
            badge_div = soup.find('div', {'data-e2e': 'rc-CreditEligibleBadge'})
            credit_eligible_text = badge_div.find('p', {'data-e2e': 'creditEligibleText'}).text.strip() if badge_div else None
            credit_eligible = bool(credit_eligible_text)
        except:
             credit_eligible = "None"


        data_dict = {
            'Partner': partner_name,
            'Course': title,
            'Skills': skills,
            'Rating': rating,
            'ReviewCount': review_count,
            'Level': level,
            'CertificateType': certificat_type,
            'Duration': duration,
            'CreditEligibility': credit_eligible
            }
        # Assuming conn is your database connection
        push_data(data_dict, conn, table_name)
```

```python
import psycopg2
import requests
from psycopg2 import sql
from selenium import webdriver
import time

def connect_to_database(dbname, user, password, host, port):
        # Establish a connection
        conn = psycopg2.connect(dbname=dbname, user=user, password=password, host=host, port=port)

        # Create a cursor object to execute SQL queries
        return conn


def push_data(data, conn, table_name):
    insert_query = sql.SQL("""INSERT INTO {} (Title, Description, Instructor, Rating, ReviewCount, Duration, Lectures, Level) VALUES 
                           (%(Title)s, %(Description)s, %(Instructor)s, %(Rating)s, %(ReviewCount)s, %(Duration)s, %(Lectures)s, %(Level)s)
                           """).format(sql.Identifier(table_name))

    with conn.cursor() as cursor:
        cursor.execute(insert_query, data)

    # Commit the changes to the database
    conn.commit()

conn = connect_to_database(dbname, user, password, host, port)
def process2(num):

    url = "https://www.udemy.com/courses/teaching-and-academics/?p={}".format(num)

    # Set up the WebDriver (make sure to have the appropriate webdriver installed)
    driver = webdriver.Chrome()

    # Open the URL
    driver.get(url)

    time.sleep(2)
    # Get the page source after JavaScript execution
    html_code = driver.page_source

    driver.quit()

    soup = BeautifulSoup(html_code, 'html.parser')

    container_div = soup.find('div', class_='filter-panel--paginated-course-list--A07TT')
    course_divs = container_div.find_all('div', class_='popper-module--popper--2BpLn')

    for soup in course_divs:
        
        try:
            course_title = soup.select_one('.course-card-title-module--title--2C6ac a').contents[0]
        except:
             course_title = "None"
        
        try:
            course_description = soup.select_one('p[data-purpose="safely-set-inner-html:course-card:course-headline"]').text
        except:
            course_description = "None"
        
        try:
            instructor = soup.select_one('div[data-purpose="safely-set-inner-html:course-card:visible-instructors"]').text
        except:
            instructional_level = "None"
        
        try:
            rating = soup.select_one('span[data-purpose="rating-number"]').text
        except:
            rating = "None"
        
        try:
            num_reviews = soup.select_one('span.course-card-ratings-module--reviews-text--1z047').text.replace(',', '').replace("(", "").replace(")", "")
        except:
            num_reviews = "None"
        
        try:
            total_hours = soup.select_one('span.course-card-details-module--row--3sv2A:nth-of-type(1)').text
        except:
            total_hours = "None"
        
        try:
            num_lectures = soup.select_one('span.course-card-details-module--row--3sv2A:nth-of-type(2)').text
        except:
            num_lectures = "None"
        
        try:
            instructional_level = soup.select_one('span.course-card-details-module--row--3sv2A:nth-of-type(3)').text
        except:
            instructional_level = "None"


        data_dict = {
            'Title': course_title,
            'Description': course_description,
            'Instructor': instructor,
            'Rating': rating,
            'ReviewCount': num_reviews,
            'Duration': total_hours,
            'Lectures': num_lectures,
            'Level': instructional_level,
            }
        # Assuming conn is your database connection
        push_data(data_dict, conn, table_name)
```
