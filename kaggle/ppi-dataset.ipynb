```python
import psycopg2
import requests
from psycopg2 import sql
from bs4 import BeautifulSoup
from datetime import datetime

dbname = "postgres"
user = ""
password = ""
host = ""
port = 5432
table_name = ""


def connect_to_database(dbname, user, password, host, port):
        # Establish a connection
        conn = psycopg2.connect(dbname=dbname, user=user, password=password, host=host, port=port)

        # Create a cursor object to execute SQL queries
        return conn


def push_data(data, conn, table_name):
    insert_query = sql.SQL("""
        INSERT INTO {} 
        (ProductName, Date, FarmPrice, AtlantaRetail, ChicagoRetail, LosAngelesRetail, NewYorkRetail, AverageSpread)
        VALUES 
        (%(ProductName)s, %(Date)s, %(FarmPrice)s, %(AtlantaRetail)s, %(ChicagoRetail)s, %(LosAngelesRetail)s, %(NewYorkRetail)s, %(AverageSpread)s)
    """).format(sql.Identifier(table_name))

    # Create a cursor and execute the insert query
    with conn.cursor() as cursor:
        cursor.execute(insert_query, data)

    # Commit the changes to the database
    conn.commit()

conn = connect_to_database(dbname, user, password, host, port)
def process1(num):

    reqUrl = "http://www.producepriceindex.com/produce-price-index?field_ppi_commodity_target_id=All&field_ppi_date_value[min]&field_ppi_date_value[max]&page={}".format(num)

    headersList = {
    "Accept": "*/*",
    "User-Agent": "Thunder Client (https://www.thunderclient.com)" 
    }

    payload = ""

    response = requests.request("GET", reqUrl, data=payload,  headers=headersList)

    soup = BeautifulSoup(response.content, 'html.parser')

    product_list = players_list = soup.find('table', class_='views-table cols-8 table').find("tbody").find_all("tr")

    for soup in product_list:

        try:
            product_name = soup.find('td', class_='fix-width-16').text.strip()
        except:
            product_name = None
        try:
            date = soup.find('td', class_='fix-width-12').find('span').text.strip()
        except:
            date = None
        try:
            farm_price = soup.find_all('td', class_='fix-width-12')[1].text.strip()
        except:
            farm_price = None
        try:
            atlanta_retail = soup.find_all('td', class_='fix-width-12')[2].text.strip()
        except:
            atlanta_retail = None
        try:
            chicago_retail = soup.find_all('td', class_='fix-width-12')[3].text.strip()
        except:
            chicago_retail = None
        try:
            los_angeles_retail = soup.find_all('td', class_='fix-width-12')[4].text.strip()
        except:
            los_angeles_retail = None
        try:
            newyork_retail = soup.find_all('td', class_='fix-width-12')[5].text.strip()
        except:
            newyork_retail = None
        try:
            average_spread  = soup.find('td', class_='views-field-expression-4').text.strip()
        except:
            average_spread = None


        data_dict = {
            'ProductName': product_name,
            'Date': date,
            'FarmPrice': farm_price,
            'AtlantaRetail': atlanta_retail,
            'ChicagoRetail': chicago_retail,
            'LosAngelesRetail':  los_angeles_retail,
            'NewYorkRetail': newyork_retail,
            'AverageSpread': average_spread
            }
        # Assuming conn is your database connection
        push_data(data_dict, conn, table_name)
        
    return len(players_list), reqUrl
```

```python
def append_word_to_file(word, filename="output.log"):
    with open(filename, 'a') as file:
        file.write(str(word) + '\n')

for i in range(146, 630):
    length, url = process1(i)
    append_word_to_file("{} - {} - {}".format(i, length, url))
    print(i)
```
